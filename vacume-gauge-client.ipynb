```python
import sys
app = QtWidgets.QApplication(sys.argv)
MainWindow = QtWidgets.QMainWindow()
ui = Ui_MainWindow()
ui.setupUi(MainWindow)
MainWindow.show()

def serialRead():
    
    ser = serial.Serial('/dev/cu.usbserial-1410')  
    sensorvalue = []
    while True:
        data = ser.readline()[:-2]
        if data:
            sensorvalue =  data.split()
            ui.lcdNumber.display(float(sensorvalue[0]))
            ui.lcdNumber_2.display(float(sensorvalue[1]))
            ui.lcdNumber_3.display(float(sensorvalue[2]))
            ui.lcdNumber_4.display(float(sensorvalue[3]))


t = threading.Thread(target = serialRead)

t.start()
sys.exit(app.exec_())
```
